```python
# relational operators

# is a < b?
a = 3
b = 10
print(a > b)

print(a == b)

print(a < b)
```

```python
# membership operator
months = ['Jan', 'Feb', 'March', 'April']
'May' in months
```

```python
# logical operators
months = ['Jan', 'Feb', 'March', 'April']
# TRUE and TRUE
print('Jan' in months and 1 < 4)

# TRUE or TRUE
print('Jan' in months or 1 < 4)

# TRUE and FALSE
print('Jan' in months and 1 > 4)

# TRUE or FALSE
print('Jan' in months or 1 < 4)

# FALSE and FALSE
print('jan' in months and 1 > 4)

# FALSE or FALSE
print('jan' in months or 1 > 4)
```

```python
# Identity Operators
a = 3
b = 3
print(a == b)
print(a is b)

a = 3
b = 3.0
print(a == b)
print(a is b)

L1 = [1,2,3]
L2 = ["ONE","TWO","THREE"]
print(L1 == L2)
print(L1 is L2)

print(L1 is not L2)
```
