Make R-style charts

Seaborn is a library for making statistical graphics in Python.
It builds on top of matplotlib and integrates closely with pandas data structures.

https://seaborn.pydata.org/tutorial.html

# types of basic seaborn plots


```python
import seaborn as sns
import pandas as pd

penguins = sns.load_dataset("penguins")

penguins
```

```python
# an example relational plot
sns.relplot(data = penguins,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'species',
           size = 'body_mass_g',
           style = 'sex')
```

```python
# an example relational plot, but 'line' kind
sns.set_style("white")
sns.relplot(data = penguins,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'species',
           size = 'body_mass_g',
           style = 'sex',
           kind = 'line')
```

```python
# an example relational plot, but 'line' kind
sns.set_style("whitegrid")
sns.lineplot(data = penguins,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'species',
           size = 'body_mass_g',
           style = 'sex')
```

```python
# An example distribution plot
sns.displot(data=penguins, 
            x="flipper_length_mm", 
            hue="species", 
            multiple="stack")

sns.displot(data=penguins, 
            x="flipper_length_mm", 
            hue="species", 
            multiple="stack", 
            kind="kde")
```

```python
sns.jointplot(data=penguins, 
              x="flipper_length_mm", 
              y="bill_length_mm", 
              hue="species")
```

```python
sns.pairplot(data=penguins, 
             hue="species")
```

```python
g = sns.PairGrid(penguins, hue="species", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

```python
# exercise

sns.histplot(data=penguins, 
            x="body_mass_g", 
            hue="sex", 
            multiple="stack")
```

```python
# exercise: ecdfplot
sns.ecdfplot(data=penguins, 
            x="body_mass_g", 
            hue="sex")
```

```python
# exercise: rugplot

sns.rugplot(data=penguins,
           x = "body_mass_g",
           y = "bill_length_mm",
           hue = "species")
sns.scatterplot(data=penguins, 
                x="body_mass_g",
                y="bill_length_mm",
                hue="species")
```

```python
# An example for categorical plot
sns.set_theme(style="ticks")
penguins = sns.load_dataset("penguins")
sns.catplot(data = penguins,
            x="species", 
            y="body_mass_g",
            hue="sex",
           #kind = "box"
           )
```

## Visualizing regression models
sns.lmplot,
sns.regplot

```python
# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=penguins,
    x="bill_length_mm", 
    y="bill_depth_mm", 
    #hue="species",
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Snoot length (mm)", "Snoot depth (mm)")
```

```python
# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=penguins,
    x="bill_length_mm", 
    y="bill_depth_mm", 
    hue="species",
    order=2
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Snoot length (mm)", "Snoot depth (mm)")
```

## Multiple panels in a figure (facets)

```python
# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=penguins,
    x="bill_length_mm", 
    y="bill_depth_mm", 
    hue="species",
    order=2,
    #col="species",
    row="species"
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Snoot length (mm)", "Snoot depth (mm)")
```

```python
g = sns.FacetGrid(data = penguins,
                  row="sex",
                  col="species", 
                  margin_titles=True)
g.map(sns.regplot, 
      "bill_length_mm", 
      "bill_depth_mm",
      color=".3", 
      fit_reg=True,
      x_jitter=0.1
     )
```
